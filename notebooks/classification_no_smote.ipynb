```python
#### Importações necessárias ####
import pandas as pd
import time
# Modelos de classificação
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# Métricas de avaliação
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
train = pd.read_pickle(r"..\data\processed\train.pkl")
test = pd.read_pickle(r"..\data\processed\test.pkl")
```

```python
x_train = train.iloc[:, 2:]
y_train = train.iloc[:, 1]

x_test = test.iloc[:, 2:] # Variáveis preditoras da base de teste dependerão da combinação selecionada
y_test = test.iloc[:, 1] # Variável alvo fixa da base de teste
```

```python
models_to_run = {
    "Random Forest": RandomForestClassifier(criterion='gini', random_state=1),
    "Logistic Regression": LogisticRegression(solver='liblinear', random_state=1)
}

all_results = [] # Armazena resultados de todos os modelos e grupos

#### Início do loop de classificação ####
for group_name, feature_list in feature_groups.items():
    print(f"\n{'='*25}\nExecutando para o grupo: {group_name}\n{'='*25}")

    #### Seleção das meta-variáveis de cada grupo ####
    x_train_atts = x_train[feature_list]
    x_test_atts = x_test[feature_list]

    #### Itera sobre cada modelo (RF e LR) ####
    for model_name, model in models_to_run.items():
        print(f"  -> Treinando {model_name}...")
        
        # Treina o modelo
        start_time = time.time()
        model.fit(x_train_atts, y_train)
        training_time = time.time() - start_time
        
        # Faz as predições e gera o relatório e matriz de confusão
        pred = model.predict(x_test_atts)
        report = classification_report(y_test, pred, output_dict=True, zero_division=0)
        cmatrix = confusion_matrix(y_test, pred)

        # Salva as métricas e matriz de confusão na lista de resultados geral
        all_results.append({
            "Grupo de Meta-Variáveis": group_name,
            "Modelo Utilizado": model_name,
            "Acurácia": report['accuracy'],
            "F1-Macro": report['macro avg']['f1-score'], ## Retorna a média da F1-macro
            "Recall-Macro": report['macro avg']['recall'], ## Retorna a média da Recall-macro
            "F1 da Classe 1 (Persuasivo)": report['1']['f1-score'], ## F1 da classe positiva
            "F1 da Classe 0 (Não-Persuasivo)": report['0']['f1-score'], ## F1 da classe negativa
            "Matriz de Confusão": cmatrix,
            "Tempo de Treinamento": training_time
        })
# ------- Fim do loop de classificação -------

# Transforma a lista dos resultados em DataFrame para melhor visualização
results_df = pd.DataFrame(all_results)
```

```python
# ------- Análise dos Resultados -------

# Ordena os resultados pelo F1-Macro em ordem decrescente e arredonda para 3 casas decimais
best_results = results_df.round(3).sort_values(by="F1-Macro", ascending=False)

print("🏆 Tabela de Classificação Final dos Modelos 🏆")
display(best_results)
```
